# Import the dataset named as Sales.csv

```python
df=pd.read_csv("Sales.csv")
```

# Explore the data useing .head() or .info() method

```python
df.head()
```

```python
df.info()
```

# Using Pivot table show total revenue of every Employee each year

```python
df['revenue']=df['Units']*df['Sales']
df.pivot_table(index="Employee", columns="Year", values="revenue", aggfunc="sum").round(0)
```

# Using pivot table show for each salesperson, the total revenue by products

```python
df['revenue']=df['Units']*df['Sales']
df.pivot_table(index="Employee", columns="Product", values="revenue", aggfunc="sum").round(0)
```

# Using pivot table show for each Location, the total revenue by products

```python
df['revenue']=df['Units']*df['Sales']
df.pivot_table(index="Product", columns="Location", values="revenue", aggfunc="sum").round(0)
```

# Using pivot table show for each Location, the total units by products

```python
df.pivot_table(index="Product", columns="Location", values="Units", aggfunc="sum")
```

# Using pivot table show for each products, the total units sold by month

```python
df.pivot_table(index="Month", columns="Product", values="Units", aggfunc="sum")
```
